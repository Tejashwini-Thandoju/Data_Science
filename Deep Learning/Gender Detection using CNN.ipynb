### Scaling 

```python
X_train.max()
```

```python
X_train_scale = X_train / 255
```

### Sequential Model 

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
```

```python
cnn_model = Sequential()
```

```python
cnn_model.add(Input(shape = (130, 130, 3)))
```

### First Convolutional Layer

```python
cnn_model.add( Conv2D(filters= 32, kernel_size=(3,3), strides = (1,1), padding = "valid", activation= "relu") )
```

```python
cnn_model.add( MaxPooling2D(pool_size=(2,2), padding = "valid"))
```

### Second Convolutional Layer

```python
cnn_model.add( Conv2D(42, kernel_size=(3,3), strides = (1,1), padding = "valid", activation = "relu"))
```

```python
cnn_model.add(MaxPooling2D(pool_size = (2,2), padding = "valid"))
```

### Fully Connected Neural Network

```python
cnn_model.add(Flatten())
```

```python
cnn_model.add( Dense(128, activation = "relu"))
```

```python
cnn_model.add( Dense(256, activation = "relu"))
```

```python
cnn_model.add( Dense(512, activation = "relu"))
```

```python
cnn_model.add( Dense(2, activation = "softmax"))
```

```python
cnn_model.summary()
```

```python
cnn_model.compile(
                    optimizer = "adam",
                    loss = "sparse_categorical_crossentropy",
                    metrics = ["accuracy"]
                    )
```

```python
cnn_model.fit(X_train_scale, y_train, epochs= 5)
```

```python
cnn_model.evaluate(X_test_scale, y_test)
```
