```python
spell=SpellChecker()
```

```python
sent1="Pyton is neww ggeneration programing languge "
```

```python
sent1
```

```python
for w in sent1.split():
    print(w)
```

```python
for w in sent1.split():
    print(w)
    c=spell.correction(word=w)
    print(c)
```

```python
wlist=[]
for w in sent1.split():
    print(w)
    c=spell.correction(word=w)
    wlist.append(c)
    res=" ".join(wlist)
    print(res)
```
