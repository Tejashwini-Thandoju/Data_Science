```python
re.findall("Red", mstr)
```

### Find Iter

```python
m = re.finditer("Red", mstr)
```

```python
m
```

```python
for i in m:
    print(i)
    print()
```

### Number Series

```python
para = """xy21zd8Aabcm322dnpZ70001abcoflo89321skYeappf632abcaxS9203caebc28KbdeNs15lwe"""
```

```python
result3 = re.findall("[0-9]",para)
print(result3)
```

```python
re.findall("[0-9][0-9]", para)
```

```python
re.findall("[0-9][0-9][0-9][0-9]", para)
```

```python
re.findall("[0-9]{4}", para)
```

```python
re.findall("[0-9]{5}", para)
```

```python
para = 'xy21zd8Aabcm32223241dnpZ70001abcof88888888lo89321skYe9182738271appf632abcaxS9203caebc28KbdeNs15lwe'
```

```python
re.findall("[0-9]{3,5}", para)
```

```python
re.findall("[0-9]{1,5}", para)
```

```python
re.findall("[0-9]{1,}", para)
```

```python
re.findall("[0-9]+", para)
```
