### Abbreviation

```python
sent3 = """AI is a futuristic technology which inclues ML , DL along with computer vision and NLP. To execute CNN and RNN model we require 
            GPU processor."""
```

```python
abbre = {
            "AI" : "Artificial Intelligence",
            "ML" : "Machine Learning",
            "DL" : "Deep Learning",
            "NLP" : "Natural Language Processing",
            "NLP." : "Natural Language Processing",
            "CNN" : "Convolutional Neural Network",
            "RNN" : "Recurrent Neural Network",
            "GPU" : "Graphical Processing Unit"
                    }
```

```python
words = sent3.split()
```

```python
words
```

```python
import time
```

```python
for w in words : 
    print(abbre.get(w, w))
```

```python
abbre.get("ak", "No Such Key")
```

```python
[abbre.get(w, w)     for w in words]
```

```python
" ".join([abbre.get(w, w)     for w in words])
```

### Stemming

```python
# Write, writes, writing... ==> write
# code, codes, coding, coder.... ==> code 

# Change, Changing, Changes, Changer, changed ==> Chang

# Universe, universal, university, universities ==> univers 
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
pstem = PorterStemmer()
```

```python
pstem.stem("writing")
```

```python
pstem.stem("running")
```

```python
pstem.stem("runner")
```

```python
pstem.stem("changing")
```

```python
pstem.stem("university")
```
