```python
inflected="""write writing writes writer written ride rider riding rides run runner running runs code coder coding codes
universe universal university universities change changes changing changer"""
```

```python
words=inflected.split()
```

```python
words
```

### PORTER STEMMER

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
pstem=PorterStemmer()
```

```python
len(words)
```

```python
import time
```

```python
for w in words:
    print("orginal :",w)
    print("Porter :",pstem.stem(w))
    print("-----------")
    time.sleep(1)
```

### SNOWBALL STEMMER

```python
from nltk.stem import SnowballStemmer
```

```python
ss=SnowballStemmer(language="english")
```

```python
ss
```

```python
ss.stem("working")
```

```python
len(words)
```

for w in words:

### LANCASTER STEMMER

```python
from nltk.stem import LancasterStemmer
```

```python
lc=LancasterStemmer()
```

```python
lc.stem("writes")
```

```python
for w in words:
    print("orginal :",w)
    print("Porter :",pstem.stem(w))
    print("Snowball :",ss.stem(w))
    print("Lancaster :",lc.stem(w))
    print("-----------")
    time.sleep(1)
```

### LAMMETAZATION

```python
import nltk
```

```python
nltk.download("wordnet")
```

```python
lemma=nltk.WordNetLemmatizer()
```

```python
lemma
```

```python
lemma.lemmatize("works")
```

```python
lemma.lemmatize("working")
```

```python
for w in words:
    print("orginal :",w)
    print("Porter :",pstem.stem(w))
    print("Snowball :",ss.stem(w))
    print("Lancaster :",lc.stem(w))
    print("Lemmatization :",lemma.lemmatize(w))
    print("-----------")
    time.sleep(1)
```
