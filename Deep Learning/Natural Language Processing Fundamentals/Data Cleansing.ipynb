###  URL Removal

```python
"www.google.com"
"https://www.kaggle.com/"
"http://www.kaggle.com/"
"https://kaggle.com/"
"http://kaggle.com/"
```

```python
sent1 = "Visit google www.google.com for performing intense web search and also vist google mail www.gmail.com for mailing server."
```

```python
sent1
```

```python
re.findall("www.\S+", sent1)
```

```python
sent2 = "Kaggle https://www.kaggle.com/ is popular http://kaggle.com/ online repository http://www.kaggle.com/ for datasets."
```

```python
re.findall("https:\S+", sent2)
```

```python
re.findall("https?:\S+", sent2)
```
