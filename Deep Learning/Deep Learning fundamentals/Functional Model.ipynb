### Functional Model

```python
ip = Input(shape = (1024,1024))
```

```python
flat = Flatten()(ip)
```

```python
den1 = Dense(1024, activation = "relu")(flat)
```

```python
den2 = Dense(512, activation = "relu")(den1)
```

```python
den3 = Dense(1000, activation = "relu")(den2)
```

```python
out = Dense(10, activation = "softmax")(den3)
```
